인덱싱:[인덱스번호]
- 0,1,2,3,..
- ..., -3, -2, -1

```python
#      0  1  2  3  4
#     -5 -4 -3 -2 -1
my = [10,20,30,40,50]
print( my[0] )
print( my[-5] )
```

### 슬라이싱
- [시작인덱스:끝인덱스:증가치]
- 시작인덱스 <=인덱스< 끝인덱스

```python
my = [10,20,30,40,50]
print( my[1:4:1] ) #1<=인덱스<4  1,2,3
print( my[1:4:2] ) #1<=인덱스<4  1,3
```

```python
my[0:5:2]# 0<=index<5 0,2,4
```

```python
my[1:4] # 증가치 생략시 1증가가 기본 my[1:4:1]
```

```python
my[:4] # 시작인덱스 생략시 0 기본 my[0:4:1]
```

```python
my[1:] # 1<=index    1,2,3,4
```

```python
### 퀴즈: 마지막데이터를 제외하고 리스트를 구하시요
my[:-1] # 0<= index < -1
```

```python
my[-3:] # -3<=index   -3,-2,-1
```

```python
my[-3::-1] #-3<= index    -3,-4,-5
```

```python
## 퀴즈  역리스트를 구하시요
my[-1::-1] #-1<= index   -1,-2,-3,...
```

```python
my[::-1]
```
