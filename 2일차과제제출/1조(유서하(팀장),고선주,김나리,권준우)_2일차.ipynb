```python
fruit = {"봄" : "딸기", "여름" : "토마토", "가을" : "사과"}
season = input('제가좋아하는계절은:')

if season in fruit :
    print('정답입니다.')
else :
    print('오답입니다.')
```

```python
score = int(input('점수 입력 : '))

if 81 <= score <= 100 :
    print('A')
elif 61 <= score <= 80 :
    print('B')
elif 41 <= score <= 60 :
    print('C')
elif 21 <= score <= 40 :
    print('D')
elif 0 <= score <= 20 :
    print('E')
```
