5. 아래와 같이 fruit 딕셔너리가 정의되어 있다. 
사용자가 입력한 값이 딕셔너리 키 (key) 값에 포함되었다면 "정답입니다"를 아닐 경우 "오답입니다" 출력하라. 
fruit = {"봄" : "딸기", "여름" : "토마토", "가을" : "사과"} 
제가좋아하는계절은: 봄 정답입니다.

````python
fruit = {"봄" : "딸기", "여름" : "토마토", "가을" : "사과"}
season = input('제가좋아하는계절은:')

if season in fruit:
  print('정답입니다.')
else :
  print('오답입니다.')
````

6. 점수 구간에 해당하는 학점이 아래와 같이 정의되어 있다. 
사용자로부터 score를 입력받아 학점을 출력하라. 
점수 학점 
81~100 A 
61~80 B 
41~60 C 
21~40 D 
0~20 E

````python
score = int(input('점수입력:'))

if 81<= score <= 100:
  print('A')
elif 61 <= score <= 80:
  print('B')
elif 61 <= score <= 80:
  print('C')
elif 41 <= score <= 60:
  print('D')
elif 21 <= score <= 0:
  print('E')
````

7. for 문을 사용해서 리스트에 저장된 값을 다음과 같이 출력하라. 
리스트 = ["김밥", "라면", "튀김"] 
오늘의 메뉴: 김밥 
오늘의 메뉴: 라면 
오늘의 메뉴: 튀김

````python
리스트 = ["김밥", "라면", "튀김"]

for n in 리스트:
  print('오늘의 메뉴:', n)
````
