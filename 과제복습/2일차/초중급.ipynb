```text
키와 몸무게를 입력받아 비만도를 구하고 결과를 출력하시요.
표준체중(kg) = (신장(cm)-100)X0.85
비만도(%) = 현재체중/표준체중(%)X100

비만도가 90%이하
저체중 ,
90 초과 ~110%
정상 ,
110 초과 ~120%
과체중 ,
120% 초과
비만

````python
신장 = float(input('신장:(신장은 100 이상입니다.)'))
현재체중 = float(input('몸무게:'))

표준체중 = (신장 - 100)*0.85
비만도 = 현재체중/표준체중*100

if 비만도 <= 90:
  print('저체중')
elif 90< 비만도 <= 110:
  print('정상')
elif 110< 비만도 <=120:
  print('과체중')
else :
  print('비만')
````
