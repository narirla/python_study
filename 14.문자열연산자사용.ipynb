문자열타입: + , * , %
- 문자열 연결: +
- 문자열 반복: *
- 포맷문자열 : %

```python
s = 'abc'
s = s+'def'
print( s )
```

```python
s = 'abc'
s = s*3
print( s )
```

```python
a = 10
b = 3.14
c = 'korea'
s = 'a=%d b=%f c=%s' %(a,b,c)
print( s )
```

```python
# 3.5버전 이후
a = 11
b = 3.14
c = 'korea'
#s = f'a={a} b={b} c={c}'
s = f'a={a:10} b={b:10.1f} c={c:>10}'
print( s )
```
