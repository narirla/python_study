1. 1cm 가 0.393701 인치인 경우 7cm가 몇인치 인지 구하시요
2. 반지름이 5이고 높이가 10인 원기둥의 부피를 구하시요
( 원기둥의부피: 원의면적X높이)
3. 폭이 10 이고 높이가 5인 사각형의 면적을 구하시요.
4. 정수가 5인 경우 짝수인지 홀수인지 구하시요.
5. fruit = ["사과", "포도", "홍시"] 에서  "홍시" 요소가 있는지 확인하시요. 결과값 '있다' 또는 '없다'

```python
cm = 7
rst = cm*0.393701
print( rst )
```

```python
r = 5
h = 10
pi = 3.14
cylinder = pi*r**2*h
print( cylinder)
```

```python
n = 5
rst = "짝수" if n%2==0 else '홀수'
print( rst )
```

```python
fruit = ["사과", "포도", "홍시"]
rst = '있다' if '홍시' in fruit else '없다'
print( rst )
```
