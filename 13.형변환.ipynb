```python
a = '100'
a = int(a) # 숫자로 변환해주는 함수
print( a )
print(type(a))
```

```python
b = '3.14'
b = float(b) # 실수로 변환해주는 함수
print( b )
print(type(b))
```

```python
c = 100
c = str(c) #문자열 변환해 주는 함수
print( c )
print(type(c))
```

```python
t = (10,20,30)
t = list(t) #리스트로 변환해 주는 함수
print(t)
print( type(t))
```
