```python
a = 5
rst = a>0 and a==5
print( rst )
```

```python
a = 5
rst = a>0 or a==0
print( rst )
```

```python
a = 5
rst =  not a>0 
print( rst )
```

```python
# 요소 in 복합데이터타입
my = [10,20,30,40,50]
rst = 100 in my
print( rst )
```

```python
# rst = 'llo' in 'hello korea'
rst = 'abc' in 'hello korea'
print( rst )
```

```python
myd = {'aa':10,'bb':20,'cc':30}
rst = 'aa' in myd  # myd.keys()  ==> ['aa','bb','cc']
print(rst)
```

```python
myd = {'aa':10,'bb':20,'cc':30}
rst = 10 in myd.values()   #[10,20,30]
print(rst)
```

```python
my = [10,20,30,40,50]
rst = 100 not in my
print( rst )
```
