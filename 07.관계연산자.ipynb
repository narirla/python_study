관계연산자(산술연사자보다 우선순위 낮다):연산 결과는  bool 타입
- >(크다)  
- >=(크거나 같다)
- <(작다)
- <=(작거나 같다)
- ==(같다)
- !=(같지않다)

```python
a = -5
rst = a>0
print( rst )
```

```python
a = -5
rst = a>=0
print( rst )
```

```python
a = 5
rst = a<0
print( rst )
```

```python
a = -5
rst = a<=0
print( rst )
```

```python
a = 3
rst = a==0
print( rst )
```

```python
a = 3
rst = a!=0
print( rst )
```
