```python
myd = {'aa':10, 'bb':20, 'cc':30}
for n in myd: #myd.keys() ['aa','bb','cc]
    print( n )
```

```python
myd = {'aa':10, 'bb':20, 'cc':30}
for n in myd.values():  #[10,20,30]
    print( n )
```

```python
myd = {'aa':10, 'bb':20, 'cc':30}
for n in myd.items():  #[('aa', 10), ('bb', 20), ('cc', 30)]
    print( n )
```
